```python
import pandas as pd
```

```python
cols=["name","survived","pclass","fare","age","sex"]
titanic=pd.read_excel("titanic3.xls",index_col="name",usecols=cols)
titanic
```

```python
titanic['sex']=="male"
```

```python
titanic[titanic['sex']=="male"]
```

```python
age_mask=titanic['age']>65
titanic[age_mask]
```

### Multiple Conditions

Use &,|,~ [not and , or ,not]

```python
female_mask=titanic["sex"]=="female"
passenger_class_mask=titanic["pclass"]==1
survival_mask=titanic["survived"]==1
titanic[female_mask & passenger_class_mask & survival_mask]
```

```python
richest_kid_mask=(titanic["age"]<20) & (titanic["fare"]>200)
rich=titanic[richest_kid_mask]
rich.nlargest(5,columns=["age","fare"])
```

```python
age_mask=titanic["age"]<21
female_mask=titanic["sex"]=="female"
survivied_mask=titanic["survived"]==1
titanic.loc[(age_mask |female_mask) & survivied_mask,"age"]
```

```python
upper_age_mask=titanic["age"]>=20
lower_age_mask=titanic["age"]<25
titanic[upper_age_mask & lower_age_mask]
```

```python
youth_mask=titanic["age"].between(20,25,inclusive="both")
```

```python
titanic[youth_mask]
```

```python
first_mask=titanic["pclass"]==1
second_mask=titanic["pclass"]==2
titanic[first_mask | second_mask]
```

```python
# or we can use
titanic[titanic["pclass"].isin([1,2])]
```
